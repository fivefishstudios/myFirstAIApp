```python
import os
import openai
import requests 
from PIL import Image 
from io import BytesIO 
from IPython.display import display

openai.api_key_path = ".api-key"
```

```python
prompt=""" 
realistic photo of an electronics engineer working with transistors and a schematic diagram on the table
"""
```

```python
response = openai.Image.create(
  prompt=prompt,
  n=2,
  # size="1024x1024"
  size="512x512"
)
```

```python
for url in response['data']:
    image_url = url['url']
    # print(image_url)
    httpResponse = requests.get(image_url)
    img = Image.open(BytesIO(httpResponse.content))
    # img.show()
    display(img)
```
